```python
import matplotlib.pyplot as plt

# 첫번째 리스트의 값은 x값, 두 번째 리스트의 값은 y로 적용
# 순서쌍(x, y)으로 매칭된 값을 좌표평면 위에 그래프 시각화
plt.plot([1, 2], [2, 4])
plt.show()
```

```python
# 하나의 숫자 리스트 입력
# 리스트의 값들이 y값들이라고 가정하고 x값 [0, 1, 2, 3]을 자동으로 만들어냄
# plt.show() 함수는 그래프를 화면에 나타나도록 함
y = [2, 3, 4, 5]
plt.plot(y)
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])
# xlabel() 함수에 'X-Label' 입력하여 x축에 대한 레이블 표시
plt.xlabel('X-Label')
# ylabel() 함수에 'y-Label' 입력하여 y축에 대한 레이블 표시
plt.ylabel('Y-Label')
plt.show()
```

```python
# plot() 함수의 label 매개변수에 'Square' 문자열 입력
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label = 'Square')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
# legend() 함수를 사용해서 그래프에 범례 표시
plt.legend()
plt.show()
```

```python
import numpy as np
# linspace() 함수로 넘파이 배열 생성
x = np.linspace(0, 10, 100)
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

# 그래프 그리기
plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = 2x')
plt.plot(x, y3, label='y = 0.5x')
plt.plot(x, y4, label='y = 3x')

# 범례 설정(ncol을 2로 설정하여 한 줄에 2개 표시)
plt.legend(loc='upper center', ncol=2)

# 타이틀 및 축 라벨 설정
plt.title('Simple Linear Plot Example')
plt.xlabel('x')
plt.ylabel('y')

# 그래프 표시
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [3, 6, 9, 12])
plt.xlim([0, 5]) # xlim[xmin, xmax]) : x축의 범위 설정
plt.ylim([0, 15]) # ylim[ymin, ymax]) : y축의 범위 설정
# 축이 증가하는 간격은 tick이 자동으로 보기좋게 설정, 따로 설정도 가능
# plt.xticks(np.arange(0,6,1))   # x축 1 간격
# plt.yticks(np.arange(0,16,3))  # y축 3 간격
plt.show()
```

```python
# 데이터 생성
x = np.linspace(1, 10, 100) # 1~10까지 범위에서 동일한 간격으로 100개의 값을 생성해 x에 저장
y = np.exp(x) # x값에 대한 지수 함수를 계산하여 y에 저장

# 선형 스케일
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Linear Scale')

# 로그 스케일
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log') # y축을 로그 스케일로 변경
plt.title('Logarithmic Scale (y-axis)')

# 로그-선형 스케일
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.xscale('log') # x축을 로그 스케일로 변경
plt.title('Log-Linear Scale (x-axis)')

# 선형-로그 스케일
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('log') # y축을 로그 스케일로
plt.xscale('linear') # x축을 선형스케일로
plt.title('Linear-Logarithmic Scale')

plt.tight_layout() # 서브 그래프 간격 조절
plt.show()
```
