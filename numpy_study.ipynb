```python
# 데이터 분석에 파이썬 리스트가 아니라 넘파이 배열을 쓰는 이유
# 1. 메모리 사용 효율이 좋고, 계산 속도가 빠름
# 2. 선형 대수 연산 지원
# 3. 벡터화 연산 지원
# 3.1 :  벡터화 연산 지원 : 배열의 모든 요소에 대해 루프 없이 한 번에 연산을 수행하는 연산 기법
# 3.2 :  요소의 값을 한번에 연산하는 요소별 연산, 값을 배열로 펼쳐서 적용하는 브로드캐스팅
```

```python
ndarray_1 = np.array([1, 2, 3]) # 넘파이 배열
ndarray_2 = np.array([4, 5, 6]) # 넘파이 배열
list_1 = [1,2,3] # 파이썬 리스트
list_2 = [4,5,6] # 파이썬 리스트

print(ndarray_1 + ndarray_2) # [5 7 9]
print(list_1 + list_2) # [1, 2, 3, 4, 5, 6]
# 넘파이 배열은 일대일 대응해서 더함 -> 요소별 연산
# 파이썬으로 넘파이 같은 연산을 할려면
# result_list = [x+y for x, y in zip(list_1, list_2)] 으로 비효율적

print(ndarray_1 * 2) # [2 4 6] = [1,2,3]*[2,2,2]
print(list_1 * 2) # [1, 2, 3, 1, 2, 3]
# 배열 간의 크기가 달라도 작은 배열을 자동으로 확장하여
# 같은 크기로 맞춘 후 연산을 수행 -> 브로드캐스팅
```

```python
a = np.array([2, 3, 4]) # 넘파이 정수형 배열 생성
print(a.dtype)
b = np.array([2.1, 3.2, 4.3]) # 넘파이 실수형 배열 생성
print(b.dtype)
c = np.array(('i','j','k')) # 파이썬 튜플로 넘파이 배열 생성
print(c.dtype)
d = np.array((True, True, False)) # 넘파이 논리형 배열 생성
print(d.dtype)
e = np.array(['ab','cd','e'], dtype='<U2') # 문자열 길이를 조절하여 배열 생성
print(e.dtype)
```

```python
# 각 학급에 4명의 학생, 각 학생이 3과목의 시험점수를 가지고 있는 배열 생성
# 예시 : 3개의 학급, 각 학급에 4명의 학생, 각 학생당 국어, 영어, 수학 점수
scores = np.random.randint(0, 101, size=(3, 4, 3))

# 학급별, 학생별, 과목별 점수를 출력
print("학급별 학생들의 점수 배열 (3차원 배열):")
print(scores)

# 학급, 학생, 과목별로 점수 출력
for class_idx in range(scores.shape[0]):
  print(f"\nClass {class_idx + 1}")
  for student_idx in range(scores.shape[1]):
    korean, english, math = scores[class_idx, student_idx]
    print(f"Student {student_idx + 1} - Korean : {korean}, English : {english}, Math : {math}")
```

```python
# zeros() 함수는 shape에 모양을 지정하여 모든 요소를 0으로 초기화
# (3, 4) 크기의 배열을 생성하여 0으로 채움
print(np.zeros((3, 4)))
# (3, 4) 크기의 배열을 데이터 타입을 정수로 생성한 후, 0으로 채움
print(np.zeros((3, 4), dtype=int))

# 모든 요소가 1로 초기화된 넘파이 배열
print(np.ones((2, 3, 4), dtype=int))

# 모든 요소가 비어 있는 넘파이 배열
print(np.empty((2, 3)))
```

```python
# arange() 함수로 넘파이 배열 만들기

# np.arange(start, stop, step, dtype)
# start: 시작값(기본:0), stop : 종료값, step: 배열사이간격(기본:1), dtype:타입(자동)
# 10 이상 30 미만까지 5 간격
print(np.arange(10, 30, 5))
# 0 이상 2 미만까지 0.3 간격
print(np.arange(start = 0, stop = 2, step = 0.3))

# linspace() 함수로 넘파이 배열 만들기
# np.linspace(start, stop, num, endpoint, retstep, dtype)
# start: 시작값, stop: 종료값, num: 배열 요소 개수(기본:50), endpoint: 종료값을 포함 여부를 나타내는 논리형 데이터(기본:True)
# retstep: 요소 사이 간격 반환 여부를 나타내는 논리형 데이터(기본:False), True로 설정하면 간격도 함께 반환
# dtype : 타입(기본:실수)
print(np.linspace(start = 1, stop = 10, num = 10, dtype = int))

# arange() 함수
# 일정 간격 강조해야하는 데이터 때 사용하기 좋음(시간, 날짜)
# 끝 범위 포함 안함
# np.arange(0, 10, 3) -> [0,3,6,9]

# linspace() 함수
# 일정한 구간 데이터 생성할때 사용
# 데이터를 시각화하거나 구간별 통계량 계산시 유용
# np.linspace(0, 10, 4) -> [0.0, 3.33, 6.67, 10.0]
```

```python
student1_scores = np.array([85,90,78])
student2_scores = np.array([92, 88, 95])

print("학생 1의 점수 : ", student1_scores)
print("학생 2의 점수 : ", student2_scores)

total_scores = student1_scores + student2_scores
print("두 학생의 총점 : ", total_scores)

average_scores = total_scores / 2
print("두 학생의 평균 점수 : ", average_scores)

student1_total = np.sum(student1_scores)
student2_total = np.sum(student2_scores)
student1_average = np.mean(student1_scores)
student2_average = np.mean(student2_scores)

print("학생 1의 총점 : ", student1_total)
print("학생 1의 평균 점수 : ", student1_average)
print("학생 2의 총점 : ", student2_total)
print("학생 2의 평균 점수 : ", student2_average)
```

```python
# reshape() 함수 : 원본 배열 변경하지 않고 새로운 배열 반환
# 데이터 순서는 유지되며 배열의 차원이나 모양만 변경
# 인수로 -1을 전달하면 자동으로 요소 개수를 맞춰줌
a= np.arange(12).reshape(3,4)
print(a)
print(a.shape)
print(a.ndim)

b = a.reshape(2,6)
print(b)
print(b.shape)

print(a)
print(a.shape)

c = a.reshape(-1, 3)
```

```python
# resieze() : 원본 배열 자체를 변형, 새 배열을 반환하지 않음
a = np.arange(12)
a.resize(4,3)
print(a)
print(a.shape)
print(a.resize(4,3))
```

```python
# ravel() : 다차원 배열을 1차원 배열로 평탄화(다차원 배열의 모든 요소를 하나의 행으로 나열)
# reshape() 함수에 -1 도 됨
a = np.arange(12).reshape(3,4)

print(a.ravel())
print(a.reshape(-1))
```

```python
# 전치 연산 : 행렬에서 행과 열을 서로 바꾸는 연산
A = np.arange(12).reshape(2,6)
print(A)

# .T : [2, 6]을 [6,2] 반환으로 전치 변환
print(A.T)
print(A.T.shape)

# 만약 전치 형태로 원본 A 배열에 저장하고 싶다면, A배열메 A.T 값을 할당
A = A.T
print(A)
print(A.shape)
```

```python
# 배열 합치기
# vstack() 수직, hstack() 수평
a = np.array([1,2,3,4]).reshape(2,2)
b = np.array([5,6,7,8]).reshape(2,2)

print(np.vstack((a,b)))
print(np.hstack((a,b)))
```

```python
# 배열 분할하기
# vsplit(), hsplit()
a = np.arange(12).reshape(2,6)
print(a)

print(np.vsplit(a,2))
print(np.hsplit(a, 3))
```
